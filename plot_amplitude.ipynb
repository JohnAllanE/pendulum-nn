```python
import os
import re

# Get the directory of the notebook
notebook_dir = os.getcwd()

# Path to the log file
log_file_path = os.path.join(notebook_dir, 'log.txt')

# Regular expression to extract Peak and Time
pattern = r"Peak:\s*(-?\d+),\s*Time:\s*(\d+)\s*ms"

# Initialize lists to store extracted values
peaks = []
times = []

# Read the log file and extract values
with open(log_file_path, 'r') as log_file:
    for line in log_file:
        match = re.search(pattern, line)
        if match:
            peak = int(match.group(1))  # Extract Peak as integer
            time = int(match.group(2))  # Extract Time as integer
            peaks.append(peak)
            times.append(time)

# Print extracted values
print("Peaks:", peaks)
print("Times:", times)
```

```python
#Sort into two lists with every even-indexed and odd-indexed peaks
even_peaks = [peaks[i] for i in range(len(peaks)) if i % 2 == 0]
odd_peaks = [peaks[i] for i in range(len(peaks)) if i % 2 != 0]
 
#Plot both even and odd peaks
import matplotlib.pyplot as plt

plt.plot(even_peaks, label='Even Peaks')
plt.plot(odd_peaks, label='Odd Peaks')
plt.xlabel('Index')
plt.ylabel('Peak Value')
plt.title('Even and Odd Peaks')
plt.legend()
plt.show()
```
